```python
import pandas as pd
```

```python
df=pd.read_csv('Data/mushrooms.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
inputs=df.drop(['class'], axis=1)
outputs=df['class']
```

```python
x=pd.get_dummies(inputs)
```

```python
x.head()
```

```python
#training_model
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
x_train, x_test, y_train, y_test= train_test_split(x, outputs, test_size=0.3)
md_1=DecisionTreeClassifier()
md_2=DecisionTreeClassifier(criterion='entropy')
md_1.fit(x, outputs)
md_2.fit(x, outputs)
```

```python
y_pred_1=md_1.predict(x_test)
y_pred_2=md_2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print('Acc_1 : %0.4f' %(accuracy_score(y_pred_1, y_test)))
```

```python
print('Acc_2 : %0.4f' %(accuracy_score(y_pred_2, y_test)))
```

```python
md_1.score(x_train, y_train)
```
