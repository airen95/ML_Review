```python
import pandas as pd
```

```python
df=pd.read_excel('Data/Iris.xls')
```

```python
df.head()
```

```python
df.columns
```

```python
inputs=df.drop(['iris'], axis=1)
outputs=df['iris']
```

```python
outputs[:5]
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT=DecisionTreeClassifier(criterion='entropy')
DT.fit(inputs, outputs)
```

```python
from sklearn import tree
from IPython.display import Image
import pydotplus
```

```python
#Visualize the tree
dot_data=tree.export_graphviz(DT, out_file=None, feature_names=inputs.columns, class_names=df.iris)
graph=pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
DT.score(inputs, outputs)
```

```python
x_new=[[4.7, 3.2, 1.3, 0.2],[6.6, 3. ,4.4, 1.4],[5.9, 3.0, 5.1, 1.8]]
```
